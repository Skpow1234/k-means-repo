# Algoritmo K-means Explicado Paso a Paso en Python

# Importar las bibliotecas necesarias

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

## Paso 1: Generar datos sintéticos para la demostración
## Vamos a crear 3 clusters de puntos de datos usando la función make_blobs de sklearn

```python
def generar_datos():
    X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)
    plt.scatter(X[:, 0], X[:, 1], s=50)
    plt.title('Datos Generados')
    plt.show()
    return X
```

## Paso 2: Aplicar el algoritmo K-means

```python
def aplicar_kmeans(X, k=3):
    # Inicializar KMeans con el número de clusters k
    kmeans = KMeans(n_clusters=k)
    
    # Ajustar los datos X al algoritmo K-means
    kmeans.fit(X)
    
    # Obtener los centros de los clusters (centroides) y las etiquetas asignadas a cada punto de datos
    centroides = kmeans.cluster_centers_
    etiquetas = kmeans.labels_
    
    # Visualizar los clusters y los centroides
    plt.scatter(X[:, 0], X[:, 1], c=etiquetas, s=50, cmap='viridis')
    plt.scatter(centroides[:, 0], centroides[:, 1], c='red', s=200, alpha=0.75, label='Centroides')
    plt.title('Resultados del Agrupamiento con K-means')
    plt.legend()
    plt.show()
    
    return centroides, etiquetas
```

## Paso 3: Explicación del algoritmo K-means

```python

def explicar_kmeans():
    explicacion = '''
    El algoritmo K-means funciona en los siguientes pasos:
    
    1. Inicialización: Se seleccionan aleatoriamente K centroides iniciales.
    2. Asignación: Se asigna cada punto de datos al centroide más cercano (basado en la distancia euclidiana).
    3. Actualización: Se recalcula el centroide de cada cluster tomando la media de los puntos asignados.
    4. Repetir: Se repiten los pasos de asignación y actualización hasta que los centroides no cambian significativamente (convergencia).
    
    Conceptos matemáticos clave:
    - La distancia entre un punto y un centroide se calcula usando la fórmula de distancia euclidiana:
    d(x, y) = sqrt(sum((x_i - y_i)^2))
    
    - El nuevo centroide se calcula como la media de todos los puntos en un cluster:
    nuevo_centroide = media(x en cluster)
    
    El objetivo es minimizar la varianza dentro del cluster (o inercia), que se define como la suma de las distancias al cuadrado
    entre los puntos y sus centroides correspondientes.
    '''
    print(explicacion)
```

```python
# Función principal para ejecutar el ejemplo
if __name__ == "__main__":
    # Paso 1: Generar datos
    X = generar_datos()
    
    # Paso 2: Aplicar K-means
    centroides, etiquetas = aplicar_kmeans(X, k=3)
    
    # Paso 3: Explicación
    explicar_kmeans()
```
